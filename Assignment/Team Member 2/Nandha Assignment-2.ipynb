```python
#Finding and replacing outliers
sns.boxplot(ff.CreditScore)
```

```python
q1=ff.CreditScore.quantile(0.25)
q3=ff.CreditScore.quantile(0.75)
print(q1)
print(q3)
```

```python
iqr=q3-q1
iqr
```

```python
ll=q1-1.5*iqr
ll
```

```python
ff.CreditScore.median()
```

```python
ff['CreditScore']=np.where(ff['CreditScore']<ll,652.0,ff['CreditScore'])
```

```python
sns.boxplot(ff.CreditScore)
```

```python
sns.boxplot(ff.NumOfProducts)
```

```python
Q1=ff.NumOfProducts.quantile(0.25)
Q3=ff.NumOfProducts.quantile(0.75)
print(Q1)
print(Q3)
```

```python
IQR=Q3-Q1
IQR
```

```python
UL=Q3+1.5*IQR
UL
```

```python
ff.NumOfProducts.median()
```

```python
ff['NumOfProducts']=np.where(ff['NumOfProducts']>UL,1.0,ff['NumOfProducts'])
```

```python
sns.boxplot(ff.NumOfProducts)
```
